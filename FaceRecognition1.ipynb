```python
import matplotlib.pyplot as plt
```

```python
from skimage.feature import hog
from skimage import data, exposure
import cv2
```

```python
image=cv2.imread('40KBphoto.jpg')
```

```python
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
fd, hog_image=hog(image, orientations=3, pixels_per_cell=(4,4), cells_per_block=(1,1), visualize=True, multichannel=True)
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(8,4), sharex=True, sharey=True)
ax1.axis('off')
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title("Input_Image")
hog_image_rescaled=exposure.rescale_intensity(hog_image, in_range=(0,20))
ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax1.set_title("Histogram od Oriented Gradients")
plt.show()
```

```python
import face_recognition
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
%matplotlib inline
```

```python
image2=cv2.imread('sl1.jpg')
face_demo2=cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3=cv2.imread('charli-square.jpg')
face_demo3=cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
image1=cv2.imread('40KBphoto.jpg')
face_demo1=cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
```

```python
sunny=face_recognition.face_encodings(face_demo2)[0]
sagar=face_recognition.face_encodings(face_demo1)[0]
charlie=face_recognition.face_encodings(face_demo3)[0]
known_face_encodings=[sunny,sagar,charlie]
```

```python
ms=cv2.imread('msd.jpg')
unknown_image=cv2.cvtColor(ms, cv2.COLOR_BGR2RGB)
plt.imshow(unknown_image)
```

```python
unknown_face_encodings=face_recognition.face_encodings(unknown_image)[0]
```

```python
cp=face_recognition.compare_faces(known_face_encodings, unknown_face_encodings)
```

```python
cp
```

```python
if cp[0]==True:
  print("It is Sunny")
elif cp[1]==True:
  print("It is Sagar")
elif cp[2]==True:
  print("It is Charlie")
else:
  print("Image NOT found")
```
